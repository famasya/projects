```python
from sklearn.linear_model import LinearRegression

# features = ['hd','ram','speed','screen','ads','trend','premium','multi']
features = ['speed','hd','ram','trend','ads']
x = train[features]
y = train.price

reg = LinearRegression()
reg.fit(x,y)
```
