```python
X = [1,2,3,4,5]
Y = [1,3,3,2,5]
```

```python
B0 = 0
B1 = 0
alpha = 0.01

terminal_error = 0.0001
index = 0

while (True):
    
    i = index % len(X)
    p = B1 * X[i] + B0
    err = p - Y[i]
    B1 = B1 - alpha * err
    B0 = Y[i] - B1 * X[i]
    
    print B1, B0
    if abs(err) < abs(terminal_error) or index is 1000:
        break
    index += 1
    
```
